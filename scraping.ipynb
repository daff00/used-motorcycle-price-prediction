Import Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
baseurl = "https://www.momotor.id" #base url

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

---
Motor Yamaha Bekas

```python
url_yamaha = []
t={}
data_yamaha =[]
c=0

for x in range(1,2): # page 1 sampai n
    k = requests.get('https://momotor.id/motor-bekas?offset=0&brands=yamaha'.format(x)).text #link page yang mau di scrape
    soup=BeautifulSoup(k,'html.parser')
    productlist = soup.find_all("div",{"class":"MuiGrid-root MuiGrid-item MuiGrid-grid-xs-6 MuiGrid-grid-sm-4 css-g8wver"}) # cari card tempat produknya

    for product in productlist: # cari link produk
        link = product.find("a", href=True)
        if link:
            url_yamaha.append(baseurl + link['href'])
```

```python
for link in url_yamaha: #input data
    f = requests.get(link, headers=headers).text
    hun = BeautifulSoup(f, 'html.parser')

    try:
        price = hun.find("p", {"class": "product-price"}).text.replace('\n', "")
    except:
        price = None
        
    try:
        location = hun.find("h2", {"class" : "product-location"}).text.replace('\n', "")
    except:
        location = None

    try:
        description = hun.find("div", {"class": "description-text"}).text.strip()
    except:
        description = None

    try:
        name = hun.find("h1", {"class": "product-name"}).text.replace('\n', "")
    except:
        name = None

    kapasitas_mesin = None
    tahun = None
    transmisi = None
    kilometer = None

    specification_list = hun.find("ul", class_="product-specification-list")

    if specification_list:
        specification_items = specification_list.find_all("li", class_="product-specification-item")

        for item in specification_items:
            label = item.find("p", class_="product-spec-label").text
            value = item.find("p", class_="product-spec-item").text

            if "Kapasitas Mesin" in label:
                kapasitas_mesin = value
            elif "Tahun" in label:
                tahun = value
            elif "Transmisi" in label:
                transmisi = value
            elif "Kilometer" in label:
                kilometer = value

    motor_yamaha = {
        "nama_motor": name,
        "lokasi" : location,
        "harga": price,
        "deskripsi": description,
        "kapasitas_mesin": kapasitas_mesin,
        "tahun": tahun,
        "transmisi": transmisi,
        "kilometer": kilometer
    }

    data_yamaha.append(motor_yamaha)
    c = c + 1
    print("completed", c)
```

```python
# Convert ke data frame
df_yamaha = pd.DataFrame(data_yamaha)
df_yamaha
```

```python
# Penyesuaian
df_yamaha['url'] = url_yamaha
df_yamaha['merk'] = 'Yamaha'
df_yamaha
```
